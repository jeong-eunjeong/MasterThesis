#1.그룹별 인원 합계

```python
df.info()
```

```python
df['대표이사, 이사, 사외이사'] = pd.to_numeric(df['대표이사, 이사, 사외이사'], errors='coerce')
df['정규직 수'] = pd.to_numeric(df['정규직 수'], errors='coerce')
df['계약직 수'] = pd.to_numeric(df['계약직 수'], errors='coerce')
df['계약직 단시간 근로자 수'] = pd.to_numeric(df['계약직 단시간 근로자 수'], errors='coerce')
```

```python
df = df.fillna(0)
df.info()
```

```python
final = df.groupby(['corp_code','성별']).agg({'대표이사, 이사, 사외이사 등':'sum','정규직 수':'sum','계약직 수':'sum','계약직 단시간 근로자 수':'sum','평균 근속 연수(년)':'mean'})
```

```python
final.head(50)
```

```python
final.to_excel(excel_writer='최종인원2023_3차.xlsx')
```
